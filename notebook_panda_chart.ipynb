```python
import pandas as pd
import requests

def load_country_data():
    """
    Load basic country data from World Bank API.
    Returns a DataFrame with country information.
    """
    try:
        # World Bank API endpoint for country data
        url = "http://api.worldbank.org/v2/country?format=json&per_page=300"

        # Make the API request
        response = requests.get(url)

        # Check if request was successful
        if response.status_code == 200:
            # Get the actual data (skipping metadata)
            data = response.json()[1]

            # Create DataFrame
            countries_df = pd.DataFrame([{
                'Country': item['name'],
                'Capital': item.get('capitalCity', ''),
                'Region': item.get('region', {}).get('value', ''),
                'Income_Level': item.get('incomeLevel', {}).get('value', ''),
                'ISO_Code': item.get('iso2Code', '')
            } for item in data])

            return countries_df

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage
if __name__ == "__main__":
    # Load the data
    df = load_country_data()

    if df is not None:
        # Display basic information
        #print("\nFirst few countries:")
        #print(df.head())

        # Display unique regions
        #print("\nAvailable regions:")
        #print(df['Region'].unique())

        # Filter for a specific region
        print("\nEast Asian countries:")
        east_asia = df[df['Region'] == 'East Asia & Pacific']
        print(east_asia[['Country', 'Capital','Region','Income_Level','ISO_Code']])

```
