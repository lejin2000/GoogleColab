```python
import yfinance as yf

def find_ticker(symbol):
    ticker = yf.Ticker(symbol)
    info = ticker.info

    # Check if valid by looking for key fields
    name = info.get("longName") or info.get("shortName")
    currency = info.get("currency")
    exchange = info.get("exchange")
    quote_type = info.get("quoteType")

    if name:
        print(f"✅ Found: {symbol}")
        print(f"   Name      : {name}")
        print(f"   Type      : {quote_type}")
        print(f"   Exchange  : {exchange}")
        print(f"   Currency  : {currency}")
    else:
        print(f"❌ Not found or no data: {symbol}")

# Try different formats for 0822EA (Principal FTSE ASEAN 40 ETF)
candidates = [
    "0822-EA.KL",
    "0822EA.KLS",
    "0822.KL",
    "CIMBA40.KL",
    "7113-WA.KL",
    "0822EA.KL"
]

for symbol in candidates:
    find_ticker(symbol)
    print()
```
